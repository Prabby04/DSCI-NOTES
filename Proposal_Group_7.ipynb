### Background

The rising level of crimes in Vancouver has raised serious concerns over the safety of Vancouverites. The police and the public are working together to count the number of crimes and find the causes. In fact, the significance of collaborative monitoring has become a key element in later explanations of criminal behavior (Koop-Monteiro & Yasmin, 2021). Andresen and Malleson (2011, as cited in Andresen et al., 2017) found fifty percent of crimes, such as assault, burglary, robbery, sexual assault, theft, and thefts of and from vehicles, over a ten year period were committed in less than nine percent of Vancouver’s streets. Most residents consider areas such as West End a lot safer than others such as Strathcona and Grandview-Woodland. Does the level of safety actually vary in different regions of Vancouver? We aim to find this out using methods of statistical inference.The Vancouver Police Department website has the data for crimes in Vancouver available for public use. We chose to focus on only one type of crime: thefts from vehicles.

### Question

### *"Concerning 2018 to 2022, is the proportion of thefts from vehicles in the Strathcona and Grandview-Woodland districts significantly lower than in the West End and Central Business districts of Vancouver, BC.?"*

# Preliminary Results

##### Read, Clean, & Wrangle

```r
#Set Seed
set.seed(2023)

# Read and Combine CSVS 
# Code from --
# Schork, J (2020). Import & Row-Bind CSV Files in R. https://statisticsglobe.com/merge-csv-files-in-r.
crime <- list.files(path = "data/", 
                       pattern = "*.csv", full.names = TRUE) %>% 
  lapply(read_csv, show_col_types=FALSE) %>%      #Quieted messages                      
  bind_rows      
head(crime)
```

##### Plot(s)

```r
crime2s <- crime |>
    select(TYPE, YEAR, NEIGHBOURHOOD) |>
    filter(TYPE == "Theft from Vehicle")

n_all_thefts <- nrow(crime2s)

crime_theft_prop <- crime2s |>
    group_by(NEIGHBOURHOOD) |>
    summarize(n_thefts = n(),
              prop = n_thefts / n_all_thefts)
crime_theft_prop
              
crime_theft_ave <- crime2s |>
    group_by(NEIGHBOURHOOD) |>
    summarize(n = n(),
              mean_theft = n / 5)
crime_theft_ave
```

```r
prop_theft_plot <- crime_theft_prop |>
    ggplot(aes(x = NEIGHBOURHOOD, y = prop, fill = NEIGHBOURHOOD)) +
    geom_bar(stat = "identity") +
    labs(title = "Proportion of Vehicle Thefts", x = "District", y = "Theft Proportions") +
    theme(text = element_text(size = 15), axis.text.x = element_text(angle = 45,hjust=1))
prop_theft_plot
```

```r
mean_theft_plot <- crime_theft_ave |>
    ggplot(aes(x = NEIGHBOURHOOD, y = mean_theft, fill = NEIGHBOURHOOD)) +
    geom_bar(stat = "identity") +
    labs(title = "Average Number of Vehicle Thefts", x = "District", y = "Average Thefts") +
    theme(text = element_text(size = 15), axis.text.x = element_text(angle = 45,hjust=1))

mean_theft_plot
```

```r
all_thefts_plot <- crime2s |>
    filter(TYPE == "Theft from Vehicle") |>
    group_by(NEIGHBOURHOOD) |>
    summarize(n = n()) |>
    ggplot(aes(x = NEIGHBOURHOOD, y = n, fill = NEIGHBOURHOOD)) +
    geom_bar(stat = "identity") +
    labs(title = "Total Number of Vehicle Thefts", x = "District", y = "Number of Thefts") +
    theme(text = element_text(size = 15), axis.text.x = element_text(angle = 45,hjust=1))
    
all_thefts_plot
```

##### Table of Estimates

```r
crime_bootstrap <- crime2s |>
                   rep_sample_n(reps =1000, size =100, replace =TRUE) |>
                   group_by(replicate, NEIGHBOURHOOD)|>
                   summarise(prop = n()/100, .groups = "keep")
head(crime_bootstrap)
tail(crime_bootstrap)
```

```r
point_estimates <- crime_bootstrap |>
                   group_by(NEIGHBOURHOOD) |>
                   summarise(mean_of_props = mean(prop))
point_estimates
```

# Methods: Plan

The above plots show the proportion of crimes, mean, and total number of vehicle thefts for each of the 4 neighbourhoods using data from years 2018-2022. The last tibble shows the proportions obtained through bootstrapping. However, in order to come to a viable conclusion, a hypothesis test must be carried out.

$H_{o}$ : The proportion of thefts from vehicles in the Strathcona and Grandview-Woodland districts **is not significantly lower** than in the West End and Central Business districts of Vancouver, BC, between 2018 to 2022.

$H_{a}$ :  The proportion of thefts from vehicles in the Strathcona and Grandview-Woodland districts **is significantly lower** than in the West End and Central Business districts of Vancouver, BC, between 2018 to 2022.

We will be comparing the proportions of these populations using a two-sample z-test:

$Z = \frac{(\hat{p_1}-\hat{p_2})-c}{((\hat{p}(1-\hat{p}))(\frac{1}{n_1}+\frac{1}{n_2}))^{1/2}}$

with $\hat{p}=\frac{n_1\hat{p_1}+n_2\hat{p_2}}{n_1+n_2}$

A **confidence** interval for the true proportion of vehicle thefts in each area group will be constructed via bootstrapping from the samples given by the VPD.

##### What do you expect to find?


We expect to find that areas locally known as "nicer" areas will have lower numbers of vehicle thefts, such as the West End, compared to neighbourhoods such as Strathcona. 

 ##### What impact could such findings have?

Determining the areas where car thefts are can aid with deciding which areas should invest more in vehicle safety security measures. 

##### What future questions could this lead to?

Once it is analyzed which areas are significantly more likely to have vehicle thefts, some other things to consider could be:

- How has the proportion and number of vehicle thefts change over the past 5, 10, and 15 years?

- Are there certain months, days, or times where vehicle thefts are at their highest?

- What vehicle type is most commonly stolen?

Overall, coming to a conclusion on vehicle theft rates and the factors behind them could assist with the prediction, prevention, and therefore lowering of total vehicle theft rates in an area.
